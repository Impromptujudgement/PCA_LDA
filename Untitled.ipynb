```python
from PIL import Image
import numpy as np
import os
```

```python
def load_images(folder_path):
    images = []
    labels = []
    for i in range(1, 41):  # 对于每个人
        for j in range(1, 11):  # 对于每个人的每张照片
            image_path = os.path.join(folder_path, f'orl{(i-1)*10+j:03d}.bmp')
            if os.path.isfile(image_path):  # 确保文件存在
                image = Image.open(image_path)
                images.append(np.array(image).flatten())  # 将图像转换为一维数组
                labels.append(i)  # 标签为人的编号
    return np.array(images), np.array(labels)
```

```python
folder_path = './data-ORL/ORL/'  # 解压后数据所在路径
images, labels = load_images(folder_path)
```

```python
train_images = images[::2]  # 奇数编号图像作为训练集
train_images = train_images.astype('float32')
train_images /= 255.0
train_labels = labels[::2]

test_images = images[1::2]  # 偶数编号图像作为测试集
test_images = test_images.astype('float32')
test_images /= 255.0
test_labels = labels[1::2]
```

```python
height, width = train_images.shape[:2]

print("Width:", width)
print("Height:", height)
```

```python
import numpy as np

def euclidean_distance(a, b):
    return np.linalg.norm(a - b)

def classify(test_sample, train_images, train_labels):
    distances = [euclidean_distance(test_sample, img) for img in train_images]
    min_distance_index = np.argmin(distances)
    return train_labels[min_distance_index]

def main():
    # 这里假设train_images, train_labels, test_images, test_labels已经被定义和初始化
    errors = np.zeros(len(np.unique(train_labels)))  # 每个类别的错误数
    total_error = 0
    print("已知类号\t样本编号\t识别结果(类号)")

    for i, (img, label) in enumerate(zip(test_images, test_labels)):
        predicted_label = classify(img, train_images, train_labels)
        error_status = "(error)" if predicted_label != label else ""
        print(f"{label}\t{i+1}\t{predicted_label} {error_status}")

        if predicted_label != label:
            errors[label - 1] += 1  # 增加对应类别的错误数
            total_error += 1

    for i, count in enumerate(errors):
        print(f"The number of errors of class {i+1} is {int(count)}")
    print(f"the number of total errors is {total_error}")
    print(f"The error rate is {total_error/len(test_labels):.2f}")
if __name__ == "__main__":
    main()
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def apply_pca_and_reconstruct(images, n_components):
    pca = PCA(n_components=n_components)
    transformed = pca.fit_transform(images)
    reconstructed = pca.inverse_transform(transformed)
    return reconstructed, pca

# 假设 train_images 是您的训练图像数据，已经被展平为一维向量
# 假设每张图像的原始尺寸是 112x92
height, width = 112, 92
n_samples = len(train_images)
train_images_reshaped = train_images.reshape(n_samples, -1)

# 选择不同的 PCA 组件数以观察重构效果
n_components_list = [10, 20,30, 50, 100,150, 200]
errors = []

for n_components in n_components_list:
    reconstructed_images, pca = apply_pca_and_reconstruct(train_images_reshaped, n_components)
    error = np.mean(np.square(train_images_reshaped - reconstructed_images))
    errors.append(error)

    # 显示第一张重构的图像作为示例
    example_image = reconstructed_images[0].reshape(height, width)
    plt.imshow(example_image, cmap='gray')
    plt.title(f"PCA with {n_components} components\nReconstruction Error: {error:.4f}")
    plt.show()

# 绘制误差图表
plt.figure()
plt.plot(n_components_list, errors, marker='o')
plt.xlabel('Number of PCA Components')
plt.ylabel('Reconstruction Error')
plt.title('PCA Reconstruction Error vs. Number of Components')
plt.show()

```

```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.neighbors import KNeighborsClassifier

# 假设 train_images, train_labels, test_images, test_labels 已经定义好
# train_images 和 test_images 是归一化后的数据，每行代表一个图像

# 应用 LDA
lda = LDA()  # 默认情况下，LDA 会使用最多 C-1 个组件，C 是类别数
train_lda = lda.fit_transform(train_images, train_labels)

# 选择未正交化的鉴别向量
# 在这里，我们不需要做额外的操作，因为 LDA.fit_transform 已经完成了这一步

# 选择正交化的鉴别向量
# 这通常需要更复杂的步骤，如格拉姆-施密特过程
# 但在许多情况下，LDA 的输出可以被认为是足够正交的

# 对测试数据应用 LDA 降维
test_lda = lda.transform(test_images)

# 使用最近邻分类器进行分类
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(train_lda, train_labels)
predicted_labels = classifier.predict(test_lda)

# 计算准确率
accuracy = np.mean(predicted_labels == test_labels)
print(f"Accuracy: {accuracy:.2f}")

# 分析错误率
errors = np.where(predicted_labels != test_labels)[0]
error_rate = len(errors) / len(test_labels)
print(f"Error Rate: {error_rate:.2f}")
```

```python
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# 假设 train_images 和 train_labels 是您的训练数据和标签

# 应用 LDA
lda = LDA()
train_lda = lda.fit_transform(train_images, train_labels)

# 获取 LDA 的鉴别向量
discriminant_vectors = lda.scalings_

# 实现格拉姆-施密特过程进行正交化
def gram_schmidt(vectors):
    orthogonal = []
    for vector in vectors.T:
        for orth in orthogonal:
            vector -= np.dot(vector, orth) * orth
        vector /= np.linalg.norm(vector)
        orthogonal.append(vector)
    return np.array(orthogonal).T

# 应用格拉姆-施密特过程
orthogonal_discriminant_vectors = gram_schmidt(discriminant_vectors)

# 把正交化后的鉴别向量应用到训练数据
train_lda_orthogonal = np.dot(train_images - lda.xbar_, orthogonal_discriminant_vectors)
test_lda_orthogonal = np.dot(test_images - lda.xbar_, orthogonal_discriminant_vectors)

# 使用分类器进行分类
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(train_lda_orthogonal, train_labels)
predicted_labels = classifier.predict(test_lda_orthogonal)

# 计算准确率
accuracy = np.mean(predicted_labels == test_labels)
print(f"Accuracy: {accuracy:.2f}")

# 分析错误率
errors = np.where(predicted_labels != test_labels)[0]
error_rate = len(errors) / len(test_labels)
print(f"Error Rate: {error_rate:.2f}")
```
